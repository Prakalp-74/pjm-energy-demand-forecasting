```python
df["Datetime"] = pd.to_datetime(df["Datetime"])
```

```python
df = df.set_index("Datetime")
```

```python
df.head()
```

```python
print("Earliest timestamp:", df.index.min())
print("Largest timestamp:", df.index.max())
df.head(3)
```

```python
plt.plot(df.index, df["PJME_MW"])
plt.title("Electricity Demand Over Time")
plt.xlabel("Date")
plt.ylabel("Demand")
plt.savefig("image/demand_over_time.png",bbox_inches="tight",dpi=300)
plt.show()
```

```python
yearly = df.groupby(df.index.year)["PJME_MW"].mean()
```

```python
yearly.plot(kind="bar")
plt.title("Average Electricity Demand by Year")
plt.xlabel("Year")
plt.ylabel("Average MW")
plt.savefig("image/yearly_trend.png", bbox_inches="tight", dpi=300)
plt.show()
```

```python
monthly = df.groupby(df.index.month)["PJME_MW"].mean()
```

```python
monthly.plot(kind="bar")
plt.title("Average Demand by Month")
plt.xlabel("Month")
plt.ylabel("Average MW")
plt.savefig("image/monthly_demand.png", bbox_inches="tight", dpi=300)
plt.show()
```

```python
hourly = df.groupby(df.index.hour)["PJME_MW"].mean()
```

```python
hourly.plot()
plt.title("Average Demand by Hour")
plt.xlabel("Hour")
plt.ylabel("Average MW")
plt.savefig("image/hourly_demand.png", bbox_inches="tight", dpi=300)
plt.show()
```

```python
df["is_weekend"] = df.index.dayofweek >= 5
```

```python
weekend = df.groupby("is_weekend")["PJME_MW"].mean()
```

```python
weekend.plot(kind="bar")
plt.title("Weekday vs Weekend Demand")
plt.xlabel("False = weekday, True = Weekend")
plt.ylabel("Average MW")
plt.savefig("image/weekday_vs_weekend.png",bbox_inches="tight",dpi=300)
plt.show()
```
